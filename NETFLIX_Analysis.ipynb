**Categorizing Vote_Average column**  :
*  Now  we have to catagorise the
Vote_Average column into popular avarage below_avg not _popular
*  we want to assign label of the Vote_Average column based on some  defined criteria .

*   For the criteria we create a function called categorize_col() function
*   function is user defind



NUMERICAL DATA TO LABEL

```python
def catagorize_col(df, col ,labels) :    #we have to put dataset for that we use df
                                         #we have to put columns for that we use col
                                         #then we have assign labels for that we use labels

   edges= [df[col].describe()['min'],    #in edges variable ,can any value be min,max &
           df[col].describe()['25%'],    #can 25%,50%,75% we write it in this & according
           df[col].describe()['50%'],    #to that we assisgn labels
           df[col].describe()['75%'],
           df[col].describe()['max']]
   df[col]= pd.cut(df[col], edges, labels=labels , duplicates = 'drop')
   return df
                                          #cut function help in  categorization
                                          #from numerical to labeel
```

```python
labels = ['not_popular', 'below_avg', 'average' , 'popular']
df = catagorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```
